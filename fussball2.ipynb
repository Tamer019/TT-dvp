```python
import pygame
import sys
import math

# Initialisierung von Pygame
pygame.init()
```

```python
# Konstanten
WINDOW_WIDTH = 800
WINDOW_HEIGHT = 600
PLAYER_RADIUS = 15
BALL_RADIUS = 10
PLAYER_COLOR = (255, 0, 0)  # Rot
BALL_COLOR = (0, 0, 255)    # Blau
PLAYER_SPEED = 5
BALL_SPEED = 10
FRICTION = 0.98  # Luftwiderstand

# Farben
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
```

```python
# Bildschirm erstellen
screen = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption('2D Fußballspiel mit Stadion')

# Laden des Hintergrundbildes
field_image = pygame.image.load('fussballfeld.jpeg')
field_image = pygame.transform.scale(field_image, (WINDOW_WIDTH, WINDOW_HEIGHT))

# Spielerposition
player_x = WINDOW_WIDTH // 2
player_y = WINDOW_HEIGHT // 2

# Ballposition und Geschwindigkeit
ball_x = WINDOW_WIDTH // 2
ball_y = WINDOW_HEIGHT // 2
ball_dx = 0
ball_dy = 0

# Schriftart
font = pygame.font.Font(None, 36)



def draw_text(text, font, color, surface, x, y):
    textobj = font.render(text, True, color)
    textrect = textobj.get_rect()
    textrect.center = (x, y)
    surface.blit(textobj, textrect)

def main_menu():
    while True:
        screen.blit(field_image, (0, 0))
        draw_text('Main Menu', font, WHITE, screen, WINDOW_WIDTH // 2, WINDOW_HEIGHT // 4)
        draw_text('Start', font, WHITE, screen, WINDOW_WIDTH // 2, WINDOW_HEIGHT // 2 - 30)
        draw_text('Settings', font, WHITE, screen, WINDOW_WIDTH // 2, WINDOW_HEIGHT // 2 + 30)
        draw_text('Exit', font, WHITE, screen, WINDOW_WIDTH // 2, WINDOW_HEIGHT // 2 + 90)

        # Spieler zeichnen
        pygame.draw.circle(screen, PLAYER_COLOR, (int(player_x), int(player_y)), PLAYER_RADIUS)

        # Ball zeichnen
        pygame.draw.circle(screen, BALL_COLOR, (int(ball_x), int(ball_y)), BALL_RADIUS)

        pygame.display.flip()

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_RETURN:
                    choose_player_color()
                if event.key == pygame.K_ESCAPE:
                    pygame.quit()
                    sys.exit()

            if event.type == pygame.MOUSEBUTTONDOWN:
                mx, my = pygame.mouse.get_pos()
                if WINDOW_WIDTH // 2 - 100 < mx < WINDOW_WIDTH // 2 + 100:
                    if WINDOW_HEIGHT // 2 - 30 - 20 < my < WINDOW_HEIGHT // 2 - 30 + 20:
                        choose_player_color()
                    elif WINDOW_HEIGHT // 2 + 30 - 20 < my < WINDOW_HEIGHT // 2 + 30 + 20:
                        pass  # Hier können Sie später Einstellungen implementieren
                    elif WINDOW_HEIGHT // 2 + 90 - 20 < my < WINDOW_HEIGHT // 2 + 90 + 20:
                        pygame.quit()
                        sys.exit()

        # Bewegung des Spielers im Menü
        keys = pygame.key.get_pressed()
        if keys[pygame.K_w]:
            player_y -= PLAYER_SPEED
        if keys[pygame.K_s]:
            player_y += PLAYER_SPEED
        if keys[pygame.K_a]:
            player_x -= PLAYER_SPEED
        if keys[pygame.K_d]:
            player_x += PLAYER_SPEED

        # Begrenzung der Spielerposition an den Fenstergrenzen
        player_x = max(PLAYER_RADIUS, min(WINDOW_WIDTH - PLAYER_RADIUS, player_x))
        player_y = max(PLAYER_RADIUS, min(WINDOW_HEIGHT - PLAYER_RADIUS, player_y))

def choose_player_color():
    global PLAYER_COLOR
    choosing = True
    while choosing:
        screen.blit(field_image, (0, 0))
        draw_text('Choose Player Color', font, WHITE, screen, WINDOW_WIDTH // 2, WINDOW_HEIGHT // 4)
        draw_text('Red', font, (255, 0, 0), screen, WINDOW_WIDTH // 2, WINDOW_HEIGHT // 2 - 30)
        draw_text('Green', font, (0, 255, 0), screen, WINDOW_WIDTH // 2, WINDOW_HEIGHT // 2 + 30)
        draw_text('Blue', font, (0, 0, 255), screen, WINDOW_WIDTH // 2, WINDOW_HEIGHT // 2 + 90)

        pygame.display.flip()

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_ESCAPE:
                    main_menu()

            if event.type == pygame.MOUSEBUTTONDOWN:
                mx, my = pygame.mouse.get_pos()
                if WINDOW_WIDTH // 2 - 100 < mx < WINDOW_WIDTH // 2 + 100:
                    if WINDOW_HEIGHT // 2 - 30 - 20 < my < WINDOW_HEIGHT // 2 - 30 + 20:
                        PLAYER_COLOR = (255, 0, 0)
                        choosing = False
                    elif WINDOW_HEIGHT // 2 + 30 - 20 < my < WINDOW_HEIGHT // 2 + 30 + 20:
                        PLAYER_COLOR = (0, 255, 0)
                        choosing = False
                    elif WINDOW_HEIGHT // 2 + 90 - 20 < my < WINDOW_HEIGHT // 2 + 90 + 20:
                        PLAYER_COLOR = (0, 0, 255)
                        choosing = False

            main_game()

    def main_game():
        global player_x, player_y, ball_x, ball_y, ball_dx, ball_dy
        running = True
        while running:
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    running = False

            # Bewegung des Spielers
            keys = pygame.key.get_pressed()
            if keys[pygame.K_w]:
                player_y -= PLAYER_SPEED
            if keys[pygame.K_s]:
                player_y += PLAYER_SPEED
            if keys[pygame.K_a]:
                player_x -= PLAYER_SPEED
            if keys[pygame.K_d]:
                player_x += PLAYER_SPEED

            # Begrenzung der Spielerposition an den Fenstergrenzen
            player_x = max(PLAYER_RADIUS, min(WINDOW_WIDTH - PLAYER_RADIUS, player_x))
            player_y = max(PLAYER_RADIUS, min(WINDOW_HEIGHT - PLAYER_RADIUS, player_y))

            # Ballbewegung
            ball_x += ball_dx
            ball_y += ball_dy

            # Ball verlangsamt sich aufgrund des Luftwiderstands
            ball_dx *= FRICTION
            ball_dy *= FRICTION

            # Ball in der Mitte respawnen, wenn er das Spielfeld verlässt
            if ball_x < BALL_RADIUS or ball_x > WINDOW_WIDTH - BALL_RADIUS or ball_y < BALL_RADIUS or ball_y > WINDOW_HEIGHT - BALL_RADIUS:
                ball_x = WINDOW_WIDTH // 2
                ball_y = WINDOW_HEIGHT // 2
                ball_dx = 0
                ball_dy = 0

            # Kollisionsprüfung Spieler-Ball
            distance = math.sqrt((player_x - ball_x) ** 2 + (player_y - ball_y) ** 2)
            if distance < PLAYER_RADIUS + BALL_RADIUS:
                # Richtung des Schusses basierend auf der Spielerbewegung
                angle = math.atan2(ball_y - player_y, ball_x - player_x)
                ball_dx = BALL_SPEED * math.cos(angle)
                ball_dy = BALL_SPEED * math.sin(angle)

            # Bildschirm mit dem Hintergrundbild füllen
            screen.blit(field_image, (0, 0))

            # Spieler und Ball zeichnen
            pygame.draw.circle(screen, PLAYER_COLOR, (int(player_x), int(player_y)), PLAYER_RADIUS)
            pygame.draw.circle(screen, BALL_COLOR, (int(ball_x), int(ball_y)), BALL_RADIUS)

            # Bildschirm aktualisieren
            pygame.display.flip()

    pygame.quit()
    sys.exit()


```

```python
if __name__ == "__main__":
    main_menu()
```
